```python
# Skybrisk Data Science Internship
# Week 2 Client Project: Data Cleaning Script

# 1. Data Structures: A list of dictionaries representing messy, raw data
raw_data = [
    {"transaction_id": 1, "customer": "Alice", "amount": 150.00},
    {"transaction_id": 2, "customer": "Bob", "amount": -50.00},   # Invalid negative amount
    {"transaction_id": 1, "customer": "Alice", "amount": 150.00}, # Duplicate entry
    {"transaction_id": 3, "customer": "Charlie", "amount": 200.00},
    {"transaction_id": 4, "customer": "David", "amount": 0.00}    # Invalid zero amount
]

print("--- Original Raw Data ---")
for row in raw_data:
    print(row)

# 2. Functions: Defining a reusable function to clean the data
def clean_dataset(data):
    cleaned_data = []
    seen_ids = set() # Using a 'Set' data structure to track unique IDs
    
    for entry in data:
        t_id = entry["transaction_id"]
        amount = entry["amount"]
        
        # Checking for and removing duplicates
        if t_id not in seen_ids:
            seen_ids.add(t_id)
            
            # Filtering data: Only keeping transactions with a positive amount
            if amount > 0:
                cleaned_data.append(entry)
                
    return cleaned_data

# Executing our custom function
processed_data = clean_dataset(raw_data)

print("\n--- Cleaned and Filtered Data ---")
for row in processed_data:
    print(row)
    
# 3. List Comprehension: A sophisticated way to extract specific data
valid_amounts = [entry["amount"] for entry in processed_data]
print(f"\nList of Valid Amounts Extracted: {valid_amounts}")
```
