```python
# Skybrisk Data Science Internship
# Week 1 Client Project: Average Temperature Calculator

# 1. Variables and Data Types (Creating a collection of daily temperatures in Celsius)
daily_temperatures = [22.5, 24.0, 19.5, 21.0, 25.5, 23.0, 20.0]

# Initializing variables to store our calculations
total_temperature = 0.0
days_recorded = 0

# 2. Loops (Using a 'for' loop to iterate through the data)
for temp in daily_temperatures:
    # 3. Operators (Adding each temperature to the total)
    total_temperature += temp  
    days_recorded += 1

# Calculating the final average
average_temperature = total_temperature / days_recorded

# 4. Input/Output (Displaying the formatted results to the screen)
print("--- Weekly Temperature Report ---")
print(f"Total days recorded: {days_recorded}")
print(f"Average Temperature: {average_temperature:.2f}°C")

# 5. Conditional Statements (Using if/else to provide an analysis of the data)
if average_temperature > 22.0:
    print("Conclusion: The data indicates a relatively warm week.")
else:
    print("Conclusion: The data indicates a relatively cool week.")
```
